```python
# Load the training and testing datasets
train_df = pd.read_csv('/content/Titanic_train.csv')
test_df = pd.read_csv('/content/Titanic_test.csv')
```

```python
# Handling missing values
# For numerical features, fill missing values with the median
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)
test_df['Age'].fillna(test_df['Age'].mean(), inplace=True)

train_df['Fare'].fillna(train_df['Fare'].mean(), inplace=True)
test_df['Fare'].fillna(test_df['Fare'].mean(), inplace=True)
```

```python
# For categorical features, fill missing values with the mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)
test_df['Embarked'].fillna(test_df['Embarked'].mode()[0], inplace=True)
```

```python
#data transformation for test
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for column in test_df.columns:
    test_df[column] = LE.fit_transform(test_df[column])
```

```python
#data transformation for train
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for column in train_df.columns:
    train_df[column] = LE.fit_transform(train_df[column])
```

```python
# Display the first few rows of the preprocessed training dataset
print("Preprocessed Training Dataset:")
print(train_df.head())

# Display the first few rows of the preprocessed testing dataset
print("\nPreprocessed Testing Dataset:")
print(test_df.head())
```

```python
X_train = train_df.drop('Survived', axis=1)
y_train = train_df['Survived']


X_test = test_df
```

```python
from sklearn.model_selection import train_test_split

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)


model = LogisticRegression(max_iter=1000)
model.fit(X_train_split, y_train_split)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Predictions on the validation set
y_val_pred = model.predict(X_val_split)
```

```python
# Evaluate performance
accuracy = accuracy_score(y_val_split, y_val_pred)
conf_matrix = confusion_matrix(y_val_split, y_val_pred)
class_report = classification_report(y_val_split, y_val_pred)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{class_report}')
```

```python
test_predictions = model.predict(X_test)


# Create a DataFrame for the predictions
submission = pd.DataFrame({'id': test_df.index, 'survived': test_predictions})
```
